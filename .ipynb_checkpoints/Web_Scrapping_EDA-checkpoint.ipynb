# Case Study - Exploratory Data Analysis (EDA) on AmbitionBox Data
### 1. Introduction:
This case study involves web scraping data from the AmbitionBox website to perform Exploratory Data Analysis (EDA). The primary goal is to analyze company reviews, employee feedback, and ratings to identify trends and insights related to the workplace environment, job satisfaction, and other metrics.

### 2. Data Collection (Web Scraping):
The data was collected by scraping company reviews from AmbitionBox using Python libraries such as BeautifulSoup and requests. The collected data includes various features like company name, ratings, review count, and more.

# Importing Necessary libraries

```python
import requests
import re
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Objective
The goal of next step was to scrape company-related data, including ratings, employee count, reviews, salaries, and other details from the AmbitionBox website. This data was collected to perform Exploratory Data Analysis (EDA) for identifying trends in various industries and understanding employee satisfaction and company characteristics.

```python
Rating = []
Company_name = []
Industry = []
Employee_count = []
Ownership = []
Age = []
Reviews = []
Salaries = []
Jobs = []
Interview =[]
Benfits = []
Headquarters = []

for i in range(1,26):
    url = f"https://www.ambitionbox.com/list-of-companies?tags=fortune500__fortune500-2022&sortBy=popular&page={i}"
    print(url)
#     print(requests.get(url))
    header = {
    'User-Agent': 'Mosszilla/5.0 (Windows NT 10.0; Win64; x64; rv:106.0) Gecko/20100101 Firefox/106.0',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.5',
    # 'Accept-Encoding': 'gzip, deflate, br',
    'DNT': '1',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1',
    'Sec-Fetch-Dest': 'document',
    'Sec-Fetch-Mode': 'navigate',
    'Sec-Fetch-Site': 'none',
    'Sec-Fetch-User': '?1',
}
    page = requests.get(url, headers=header)

#     print('Accessing response number is:',page.status_code)
    
    soup = BeautifulSoup(page.text)
    
    Container1 = soup.find_all('div', class_='companyCardWrapper')
    Container2 = soup.find_all('span',class_ = 'companyCardWrapper__interLinking')
    

        # Extract the data from each container
    for container in Container1:
        # Rating
        rating = container.find('span', class_='companyCardWrapper__companyRatingValue')
        if rating:
            Rating.append(rating.text.strip())
        else:
            Rating.append(np.nan)

    for i in soup.find_all('h2',class_ = 'companyCardWrapper__companyName'):
        Company_name.append(i.text.strip())
        
    for i in Container2:
        text = i.text
        
        # Industry name
#         rem_special = lambda s: re.sub(r'[^a-zA-Z0-9\s]','',s).text.strip()
        I_name = re.findall('^[\w\s&]+',text.strip())
        if I_name:
            Industry.append(I_name[0])
        else:
            Industry.append(np.nan)

        Hq = re.findall(r'\|\s([A-Za-z\/]+)\s\+\d+\smore',text.strip())
        if Hq:
            Headquarters.append(Hq[0])
        else:
            Headquarters.append(np.nan)
            
#         Employee_count
        E_count = re.findall("(\d+\s*(?:k|Lakh)?\s*(?:-\s*\d+\s*(?:k|Lakh)?)?\s*Employees?)|(\d+)(?:\s*(?:Lakh)?\s*\+?\s*Employees?) ",text.strip())
#         rem_special = lambda s: re.sub(r'[^a-zA-Z0-9\s]','',s).text.strip()
#         special = rem_special(E_count)
        if E_count:
            Employee_count.append(E_count[0])
        else:
            Employee_count.append(np.nan)
            
        # ownership
        sector_name = re.findall('(Public|Fortune India 500)',text.strip())
        if sector_name:
            Ownership.append(sector_name[0])
        else:
            Ownership.append(np.nan)
            
        # Age
        age_old = re.findall(r'\s\d+\syears\sold',text.strip())
        if age_old:
            Age.append(age_old[0])
        else:
            Age.append(np.nan)


    for i in soup.find_all('div', class_='companyCardWrapper'):
        
        # Reviews
        review_count = i.find('a', href = re.compile('reviews')).find('span', class_='companyCardWrapper__ActionCount')
        if review_count:    
            Reviews.append(review_count.text.strip())
        else:
            Reviews.append(np.nan)
        
        # Salaries
        salary_count = i.find('a', href = re.compile('salaries')).find('span', class_='companyCardWrapper__ActionCount')
        if salary_count: 
            Salaries.append(salary_count.text.strip())
        else:
            Salaries.append(np.nan)
            
        # Jobs
        job_count = i.find('a', href = re.compile('jobs')).find('span', class_='companyCardWrapper__ActionCount')                
        if job_count:
            Jobs.append(job_count.text.strip())
        else:
            Jobs.append(np.nan)
            
        # Interviews    
        interview_count = i.find('a', href = re.compile('interviews')).find('span',class_ = 'companyCardWrapper__ActionCount')
        if interview_count:
            Interview.append(interview_count.text.strip())
        else:
            Interview.append(np.nan)
        
        # Benefits
        Bene_count = i.find('a', href = re.compile('benefits')).find('span', class_ = 'companyCardWrapper__ActionCount')
        if Bene_count:
            Benfits.append(Bene_count.text.strip())
        else:
            Benfits.append(np.nan)
```

```python
dict_ = {'Company':Company_name,'Rating':Rating,'Industry':Industry,"Employees":Employee_count,"Location":Headquarters,"Ownership":Ownership,"age":Age,'Reviews':Reviews,'Salaries':Salaries,'Jobs':Jobs,'Interviews':Interview,'Benefits':Benfits}
import pandas as pd
df = pd.DataFrame(dict_)
```

# Now filling nan values
Now in this step we are going to cconvert 'k' and 'L' to numbers

```python
def Conversion(X):
    A = []
    for c in X:
        A.append(c)
    if A[-1].isalpha:
        B = re.findall("\d+",X)
        if B:
            result = int(''.join(B))
            if A[-1]=="k":
                Actual = (result)*100
                return Actual
            elif A[-1]=="L":
                Actual = (result)*10000
                return Actual
            else:
                return result
        else:
            return 0
```

```python
# Conversion of Reviews
df['Reviews'] = [Conversion(x) for x in df['Reviews']]

# Conversion of Interviews
df['Interviews'] = [Conversion(x) for x in df['Interviews']]

# Conversion of job
df["Jobs"] = [Conversion(x) for x in df["Jobs"]]

# Convertion of Interview 
df["Salaries"] = [Conversion(x) for x in df["Salaries"]]

# conversion of benfits
df["Benefits"] = [Conversion(x) for x in df["Benefits"]] 
```

```python
def process_employee_count(value):
    if pd.isna(value):
        return np.nan
    if value == ('', '1'):
        return '100000+'
    elif value == ('50k-1 Lakh Employees', ''):
        return '50000-100000'
    elif value == ('10k-50k Employees', ''):
        return '10000-50000'
    elif value == ('5k-10k Employees', ''):
        return '5000-10000'
    elif value == ('1k-5k Employees', ''):
        return '1000-5000'
    elif value == ('201-500 Employees', ''):
        return '201-500'
    elif value == ('501-1k Employees', ''):
        return '501-1000'
    elif value == ('1-10 Employees', ''):
        return '1-10'
    elif value == ('51-200 Employees', ''):
        return '51-200'
    elif value == ('11-50 Employees', ''):
        return '11-50'
    else:
        return value

df['Employees'] = df['Employees'].apply(process_employee_count)
```
